```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
i = 1
df = {}
fle_name = {}

for fle in glob.glob("data/eurusd/*.csv"):
    df[i] = pd.read_csv(fle)
    df[i]['round_pls'] = np.round(df[i]['pls'],4)
    fle_name[i] = fle
    i += 1
```

```python
new_df = pd.DataFrame()

for j in range(1,i):
    new_df[j] = df[j].groupby(['date_val'])['pls'].sum()
    print(f'{fle_name[j]} : {np.sum(new_df[j])}')
new_df
```

```python
new_df = pd.DataFrame()

for j in range(1,i):
    new_df[j] = df[j].groupby(['close_type'])['pls'].sum()
    print(f'{fle_name[j]} : {np.sum(new_df[j])}')
new_df
```

```python
new_df = pd.DataFrame()

for j in range(1,i):
    new_df[j] = df[j].groupby(['close_type','round_pls'])['pls'].count()
new_df
```
