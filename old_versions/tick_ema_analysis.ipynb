```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/yearly_tick_data/2020.csv', nrows = 100_000)

df['tick'] = (df['Bid'] + df['Ask'])/2
df = df[['tick']]
df['ema'] = df['tick'].ewm(span = 600).mean()
df['tick_diff'] = df['tick'].diff(-600)
df['ema_diff'] = df['ema'].diff(-600)

df = df.dropna()
df.describe()
```

```python
df = pd.read_csv('data/yearly_tick_data/2020.csv', nrows = 100_000)

df['tick'] = (df['Bid'] + df['Ask'])/2
df = df[['tick']]
df['ema'] = df['tick'].ewm(span = 10).mean()
df['tick_diff'] = df['tick'].diff(-600)
df['ema_diff'] = df['ema'].diff(-600)

df = df.dropna()
df.describe()
```

```python
figure(num=None, figsize=(50, 10), dpi=80, facecolor='w', edgecolor='k')

df['tick'].plot(label = 'tick')
df['ema'].plot(label = 'ema')
plt.legend(loc="upper left")
```
